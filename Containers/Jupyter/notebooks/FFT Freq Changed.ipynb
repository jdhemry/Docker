```python
import matplotlib.pyplot as plt
import numpy as np

np.set_printoptions(threshold=np.inf)

plt.style.use('seaborn-poster')
%matplotlib inline

# sampling rate
sr = 2048
# sampling interval
ts = 2.0/sr
t = np.arange(0,2,ts)

freq = 1.
x = 3*np.sin(2*np.pi*freq*t)

freq = 4
x += np.sin(2*np.pi*freq*t)

freq = 7   
x += 0.5* np.sin(2*np.pi*freq*t)
x *= 200;

#print(x)
#y = map(lambda z: z * 200, x)
#print(y)
x_int = x.astype(int)
#print(x_int)
#print(*x_int, sep = ", ")

plt.figure(figsize = (8, 6))
plt.plot(t, x_int, 'r')
plt.ylabel('Amplitude')

plt.show()
```

```python
from numpy.fft import fft, ifft

sr2 = 2048

#X = fft(x_int)
s = slice(sr2)
x_2 = x_int[s];
# print(x_2)
X = fft(x_2)
# print(X)

N = len(X)
n = np.arange(N)
T = N/sr2
freq = n/T 

plt.figure(figsize = (12, 6))
plt.subplot(121)

plt.stem(freq, np.abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(0, 10)

# plt.subplot(122)
# plt.plot(t, ifft(X), 'r')
# plt.xlabel('Time (s)')
# plt.ylabel('Amplitude')
# plt.tight_layout()
plt.show()
```
